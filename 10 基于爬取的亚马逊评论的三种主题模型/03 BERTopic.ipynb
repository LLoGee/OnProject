# 3. train BERTopic

```python
from bertopic import BERTopic

# topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True)
topic_model = BERTopic(language="english", verbose=True)
topics, probs = topic_model.fit_transform(data_words)
```

# 4. 保存模型与抽取的结果

```python
import pickle

BERT_topic = topic_model.get_topic_info().head(100).to_excel(f'Data/BERT_topics.xlsx')
pickle.dump(topic_model.get_topic_info(), open(f'Data/Full_BERT_topics.save', 'wb'))
pickle.dump(topic_model, open(f'Models/BERTopic_model.model', 'wb'))
```
