```python
import pandas as pd
import matplotlib.pyplot as plt
sentence_lengths = list(pd.read_csv('./data/sentence_lengths.csv')['lengths'])
```

```python
plt.figure(figsize=(8, 5))
plt.hist(sentence_lengths, bins=15, edgecolor='black')  # bins参数表示直方图的条形数

# 添加标签和标题
plt.xlabel('lengths')
plt.ylabel('occurrences')
plt.title('Data length histogram')

# 显示直方图
plt.show()
```

```python
class_counts = pd.read_csv('./data/Data/train_data_label_distribution.csv')['count'].tolist()
```

```python
min(class_counts)
```

```python
plt.figure(figsize=(6, 4.3))
plt.plot(class_counts[1: ])  # bins参数表示直方图的条形数

# 添加标签和标题
plt.xlabel('class')
plt.ylabel('counts')
plt.title('Positive Sample frequency of different types of events')
# 显示直方图
plt.show()
```

```python
p = 100 * class_counts[0] / sum(class_counts)
p =  round(p, 2)
sizes = [p, round(100-p,2)]  # 数据大小
labels = ['negative samples', 'postive samples']  # 标签

# 绘制饼状图
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# 添加标题
plt.title('The ratio of positive and negative samples')

# 显示图表
plt.axis('equal')  # 使饼状图比例保持一致
plt.show()
```

```python
train_percentages = [(num / sum(class_counts[1: ])) * 100 for num in class_counts[1: ]]
test_class_counts = pd.read_csv('./data/Data/valid_data_label_distribution.csv')['count'].tolist()
test_percentages = [(num / sum(test_class_counts[1: ])) * 100 for num in test_class_counts[1: ]]
```

```python
plt.figure(figsize=(8,5))
plt.plot(train_percentages, label='Train', marker='o')


# 绘制第二个列表的折线
plt.plot(test_percentages, label='Valid', marker='x', linestyle='--' )

# plt.title('折线图示例')

# 添加图例
plt.legend()

# 添加轴标签
plt.xlabel('Class')
plt.ylabel('Class Sample Percentage % ')

# 显示图形
plt.show()
```
