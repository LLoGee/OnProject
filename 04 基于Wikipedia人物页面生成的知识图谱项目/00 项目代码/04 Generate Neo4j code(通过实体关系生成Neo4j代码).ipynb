```python
# 2. Load Entities and Relations
import pandas as pd
Ens = pd.read_csv("Data/Entities_Relations/" + "Rebel_Entities.csv")
Rls = pd.read_csv("Data/Entities_Relations/" + "Rebel_Relations.csv")
```

```python
Ens_dic = {}
Ens_Code_str = ""
Return_Code_str = "RETURN "
# 1. Create entity dictionary
# and complete the code writing
for i in range(len(Ens["0"])):
  Ens_dic.update({Ens["0"][i]: 'e'+str(i)})
  Return_Code_str = Return_Code_str + 'e'+str(i)+','
  Ens_Code_str = Ens_Code_str + "CREATE (" + Ens_dic[Ens["0"][i]] + ":Define_manual {title: " + '\"' + Ens["0"][i]+ '\"' + "})" + "\n"
Return_Code_str = Return_Code_str[:-1]

# 2. Then complete the relationship code writing
# CREATE (t1)-[r1:poetry_author]->(a1)
# By understanding the code, first find the code corresponding to the head and tail entities,
# Use the dictionary Ens_dic[Rls["head"][i]] Ens_dic[Rls["tail"][i]]
# The above representation, the head entity code and tail entity code corresponding to the i-th relationship
# The middle r1 seems to be unnecessary, it is necessary for: poetry_author,
# Therefore, the i-th relationship formula is expressed as CREATE (Ens_dic[Rls["head"]][i]])-[:Rls["type"][i]]->(Ens_dic[Rls["tail"][ i]])

Rls_Code_str = ""
for i in range(len(Rls["head"])):
  Relation = Rls["type"][i].replace(" ","_")
  Rls_Code_str = Rls_Code_str + "CREATE (" + Ens_dic[Rls["head"][i]] + ")-[:"+ Relation +"]->("+ Ens_dic[Rls["tail"][i]] + ")" + "\n"
```

```python
# 3. Finally merge all the codes into one txt file
All_Code_str = Ens_Code_str + Rls_Code_str + Return_Code_str
with open("Data/Neo4j_Code/Stage_one.txt", "w+") as file:
  file.write(All_Code_str)
  file.close()
```
