## 4. Select 选择框
       adio框及checkbox框都是input元素，只是里面的type不同而已。

       select框 则是一个新的select标签，大家可以对照浏览器网页内容查看一下

           对于Select 选择框， Selenium 专门提供了一个 Select类 进行操作
           
    

### 1. 根据 value属性进行选择，那么此时可能就得获取所有 select元素 的 option 子元素，然后获取value元素值
    s.select_by_value('foo')
### 2. 根据 元素的 text 文本内容进行选择
    s.select_by_visible_text('Bar')
### 3. 根据 子元素出现的次序进行选择
    select_by_value
    deselect_by_index
    deselect_by_visible_text
    deselect_all
    
    等等方法，建议查阅文档了解所有方法

三种方法都可以使用，但最常使用的是根据元素的可见文本进行选择

    注意需要导入 selector 类， 并且需要对element进行装载，然后根据select方法进行选择
        而并非直接根据 元素方法
    
    对于多选问题，我们一般先去除所有已经选择的对象也就是元素 select.deselect_all()
        然后进行选择
